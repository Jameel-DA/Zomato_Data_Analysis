```python
# change column rating
df['rate'] = df['rate'].str.replace('/5','').astype(float)
```

```python
df['rate'].dtype
```

Type of restaurant

```python
# What type of restaurant do the majority of customers order from?
ax = sns.countplot(x=df['listed_in(type)'],data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Restaurant")
plt.show()
```

conclusion - Dining Restauratns are preferred by a large number of individuals. 

```python
# How many votes has each type of restaurant received from customers?
group_data = df.groupby('listed_in(type)')['votes'].sum()
plt.plot(group_data, data=df, marker='d', c='g')
plt.xlabel("Types of Restaurant")
plt.ylabel("Votes")
plt.show()
```

The majority of restaurants recieved rating

```python
# What are the ratings that the majority of restaurants have recieved?
sns.displot(x=df['rate'],data=df,bins=10)
plt.title("Raiting Distribution")
plt.show()
```

The majority of couples prefer restaurants with an approxiamate cost of 300 rupees

```python
# Same Plot create Using matplotlib
plt.hist(df['rate'],bins=5)
plt.title("Raiting Distribution")
plt.show()
```

```python
# Zomato has Observed that most couples order most of their food online. What is their average spending on each order?
sns.countplot(x=df['approx_cost(for two people)'],data=df,palette='PuOr')
plt.show()
```

Whether online orders recieve higher ratings than offline orders 

```python
# Which Mode (online or offline) has recieved the maximum rating?
ax = sns.countplot(x=df['online_order'],data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

Offline order recieve lower rating than higher than online orders 

```python
# Same Question solve using box plot
sns.boxplot(x=df['online_order'],y=df['rate'],data=df)
plt.show()
```
